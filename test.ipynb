```python
import jieba  
from collections import Counter  
from wordcloud import WordCloud  
import matplotlib.pyplot as plt  
  
#   
text = ""


# jieba  
words = jieba.lcut(text)  
  
# collectionsCounter  
word_counts = Counter(words)  
  
#   
stopwords = set()
content = [line.strip() for line in open('../cn_stopwords.txt','r').readlines()] + ["","","","","","","","","","","","","","","", "", "", "", "", "","","","","","","","","","","","","","","","","","","",""]
stopwords.update(content)
# stopwords.add(["","","","","","","","","","","","","","","", "", "", "", "", "","","","","","","","","","","","","","","","","","","",""])  #   
word_counts = {word: count for word, count in word_counts.items() if word not in stopwords}  
  
#   
wc = WordCloud(font_path='C:/Windows/Fonts/simhei.ttf',  #   
               background_color="white",  #   
               max_words=100,  #   
               width=3200,  #   
               height=2400,  #   
               random_state=42  #   
               )  
wordcloud = wc.generate_from_frequencies(word_counts)  
  
# matplotlib  
plt.figure(figsize=(10, 8))  
plt.imshow(wordcloud, interpolation='bilinear')  
plt.axis("off")  
plt.show()
```
